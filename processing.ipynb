## Data Processing

This notebook performs the processing required to transform raw data into usable format for training the model.

In this case, a tabular format that represents the raw textual data to be used as input for the models as well as encoding the targets as outputs for the model.

```python
import pandas as pd
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from sentence_transformers import SentenceTransformer
```

```python
data = pd.read_csv('data/clean/cleaned_data.csv')
data
```

Convert the labels into target binary vector

```python
target_enc = LabelBinarizer()
y = target_enc.fit_transform(data.target.tolist())
```

```python
target_enc.classes_ # same order as index
```

```python
y = pd.DataFrame(y, columns=target_enc.classes_)
y
```

```python
y.to_parquet('data/processed/target.parquet', index=False)
```

Extract sentence embeddings as features using Sentence Transformer

```python
st = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
X = st.encode(data['statement'].tolist(), convert_to_numpy=True)
X
```

```python
X_df = pd.DataFrame(X)
```

```python
X_df
```

```python
X_df.columns = [str(c) for c in X_df.columns]
X_df.to_parquet('data/processed/baseline_0_features.parquet')
```
